```python
from dotenv import load_dotenv
import requests
import json
import os

# Send GET request to fetch available models
load_dotenv() 
ELEVENLABS_API_KEY = os.getenv("ELEVENLABS_API_KEY")

resp = requests.get(
    "https://api.elevenlabs.io/v1/models",
    headers={"xi-api-key": ELEVENLABS_API_KEY}
)

# Parse and pretty-print the JSON response
models = resp.json()
print(json.dumps(models, indent=2, ensure_ascii=False))
```

```python
import requests
import json
import os
from dotenv import load_dotenv

load_dotenv()
ELEVENLABS_API_KEY = os.getenv("ELEVENLABS_API_KEY")
voice_id = os.getenv("ELEVENLABS_VOICE_ID")  # Default voice ID

# === CONFIGURATION ===
 # Replace with your actual key
VOICE_ID = voice_id  # Replace with your voice ID

# === API REQUEST ===
url = f"https://api.elevenlabs.io/v1/voices/{VOICE_ID}/settings"

headers = {
    "xi-api-key": ELEVENLABS_API_KEY,
}

response = requests.get(url, headers=headers)

# === HANDLE RESPONSE ===
if response.status_code == 200:
    settings = response.json()
    print(json.dumps(settings, indent=2))  # Nicely formatted output
else:
    print(f"❌ Error {response.status_code}: {response.text}")
```
